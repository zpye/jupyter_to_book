```python
#format
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def _to_latex_(self):
    str = self.to_latex(escape = False, 
                        column_format = '@{{}} *{{{0}}}{{c}} @{{}}'.format(self.columns.size + 1))
    return r'''\begin{{table}}[!h]
    \begin{{center}}
    {0}
    \end{{center}}
    \end{{table}}'''.format(str)

pd.DataFrame._repr_latex_ = _to_latex_

import book_format
style = book_format.set_style()
```

## matplotlib

```python
t = np.arange(0, 10, 0.01)
s1 = np.sin(2 * np.pi * t)
s2 = np.sin(2 * np.pi * 1.5 * t +  0.3 * np.pi) - 0.3

(fig, ax) = plt.subplots(dpi = 300)
ax.plot(t, s1, t, s2)
ax.set_xlabel('time')
ax.set_ylabel('value of s1 and s2')
ax.grid(True)

fig.tight_layout()
plt.show()
```
